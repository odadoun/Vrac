```python
import sys
sys.path.insert(1, '/home/beau/Git/pycoa')   #adapt the path for local install, if necessary. Under google colab, do not change anything
import coa.geo as cg
f=cg.GeoCountry('FRA')
f.set_dense_geometry()
z=f.get_data()
display(z)
```

```python
z['leet_name']=z.name_subregion.str.upper().str.normalize('NFKD').str.encode('ascii',errors='ignore').str.decode('utf-8') # plus d'accents

d={'A':'4','B':'8','E':'3','G':'6','I':'!','L':'1','O':'0','R':'2','S':'5','T':'7'}
for k,v in d.items():
    z['leet_name']=z.leet_name.str.replace(k,v)
```

```python
import geopandas as gpd
```

```python
ax=z.plot(figsize=(16,16))
for idx, row in z.iterrows():
    ax.annotate(s=row.leet_name, xy=row.geometry.centroid.coords[0], ha='center', va='center', size=8)#, weight='bold')
```
